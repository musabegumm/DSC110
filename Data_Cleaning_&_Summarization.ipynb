# Data Cleaning

Down below, we have "isna" where it counts or looks for any missing values and when we say "sum" it counts how many values each columns has.

df.isna().sum

```python
df.isna().sum()
```

Down here, its calculating the missing values for each columns.

```python
df.isna().mean()*300
```

Down here, the code will remove or clean the row for "systolic_bp"

```python
df_clean = df.dropna(subset=['systolic_bp']).copy()
```

```python
df_clean.isna().mean()*100
```
