```python

#One Numerical Variable
df['bmi'].hist(color='steelblue', edgecolor='black')
plt.title('Distribution of BMI')
plt.xlabel('BMI')
plt.ylabel('Number of Respondents')
plt.grid(False)
```

Violin Charts

Heatmap

```python
import seaborn as sns

# Filter to only Yes (1) and No (2) diabetes responses
df_diabetes = df[df['has_diabetes'].isin([1, 2])].copy()

diabetes_map={1:'Yes', 2:'No'}
df_diabetes['has_diabetes']=df_diabetes['has_diabetes'].map(diabetes_map)

# Create violin plot
sns.violinplot(data=df_diabetes, x='has_diabetes', y='bmi')
plt.title('BMI Distribution by Diabetes Status')
plt.xlabel('Diabetes Status (1=Yes, 2=No)')
plt.ylabel('BMI (kg/m²)')
```

Customizing the visulization

plt shot for matplotlib --> which is creating wide variety of different type of visulization.   The x-axis tralstes to ages. While the y-axis is translating the number of respondents.

"grid(False)" --> removes the grid lines from the visulization

```python
df['age'].hist(color='purple' , edgecolor='black' , bins=7)
plt.title('Age Distribution among NHANES Survey Respondents')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(False)
```

```python
# Filter to only Yes (1) and No (2) diabetes responses
df_diabetes = df[df['has_diabetes'].isin([1, 2])].copy()

diabetes_map={1:'Yes', 2:'No'}
df_diabetes['has_diabetes']=df_diabetes['has_diabetes'].map(diabetes_map)

# Create violin plot
sns.violinplot(data=df_diabetes, x='has_diabetes', y='bmi')
plt.title('BMI Distribution by Diabetes Status')
plt.xlabel('Diabetes Status (1=Yes, 2=No)')
plt.ylabel('BMI (kg/m²)')
```

Downbelow, we're creating a list of all the columns we have in the dataframe

```python
# Select only numeric columns
numeric_cols = df[['age', 'bmi', 'systolic_bp', 'hba1c', 'fasting_glucose']].corr()

sns.heatmap(numeric_cols, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Between Numeric Variables')
```
