```python
#Mapping labels, here we're translating the numbers into words
gender_map     = {1.0: 'Male', 2.0: 'Female'}
race_map       = {1.0: 'Mexican American', 2.0: 'Other Hispanic',
                  3.0: 'White', 4.0: 'Black',
                  6.0: 'Asian', 7.0: 'Other'}
education_map  = {1.0: 'Less than 9th', 2.0: '9-11th grade',
                  3.0: 'High school',   4.0: 'Some college',
                  5.0: 'College grad',  9.0: 'Do not know'}

df['gender']         = df['gender'].map(gender_map)
df['race_ethnicity'] = df['race_ethnicity'].map(race_map)
df['education']      = df['education'].map(education_map)
```

```python
#Dictionaries, the code here converts the numerical into readable texts
gender_map     = {1.0: 'Male', 2.0: 'Female'}
race_map       = {1.0: 'Mexican American', 2.0: 'Other Hispanic',
                  3.0: 'White', 4.0: 'Black',
                  6.0: 'Asian', 7.0: 'Other'}
education_map  = {1.0: 'Less than 9th', 2.0: '9-11th grade',
                  3.0: 'High school',   4.0: 'Some college',
                  5.0: 'College grad',  9.0: 'Do not know'}

df['gender']         = df['gender'].map(gender_map)
df['race_ethnicity'] = df['race_ethnicity'].map(race_map)
df['education']      = df['education'].map(education_map)
```

```python
#Checking for missing values
df.isna().sum()
```

```python
#This code here fills in missing BMI with average BMI
df['bmi'] = df['bmi'].fillna(df['bmi'].mean())
```

```python
df.isna().sum()
```

```python
#This also fills the missing values of height_cm and systolic_bp with their average numbers
df['height_cm'] = df['height_cm'].fillna(df['height_cm'].mean())
df['systolic_bp'] = df['systolic_bp'].fillna(df['systolic_bp'].mean())
```

```python
df.isna().sum()
```

```python
#It removes all rows where the education is missing
df = df.dropna(subset=['education']).copy()
```

# One Hot Encodoing

```python
#This code convers the categorical columns into numerical columns
df = pd.get_dummies(df, columns=['gender', 'race_ethnicity', 'education'],
                    drop_first=True)
```

```python
df.head()
```

## Scaling Numerical Numbers

```python
#Scaling the numerical columns
numeric_cols = ['age', 'bmi', 'height_cm', 'systolic_bp']
scaler = StandardScaler()
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])
```
